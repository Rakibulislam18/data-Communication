```python
import numpy as np
import matplotlib.pyplot as plt

def analog_signal(time, a, f, delta):
    return a*np.sin(2*np.pi*f*time + delta)

f = 4
delta = 0
data = [1, 0, 1, 1, 0, 0, 1,0]
t = np.arange(0, len(data), .001)
y = np.array([])

start = 0
size = 1000
for x in data:
    if x == 1:
        y = np.append(y, analog_signal(time=t[start: start+size], a = 2, f= f, delta= delta))
        

    if x == 0:
        y = np.append(y, analog_signal(time=t[start: start+size], a = .25, f= f, delta= delta))

plt.plot(t, y)
plt.show()
```
